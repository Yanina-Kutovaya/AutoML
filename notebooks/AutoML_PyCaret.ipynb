```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from pycaret.regression import *
```

## 1. Data ingestion

```python
#!pip install -q kaggle
#!mkdir ~/.kaggle
#!cp kaggle.json ~/.kaggle/
#!chmod 600 ~/.kaggle/kaggle.json
#!kaggle competitions download -c mercedes-benz-greener-manufacturing
#!unzip mercedes-benz-greener-manufacturing.zip
```

```python
train = pd.read_csv('train.csv.zip', index_col='ID')
test = pd.read_csv('test.csv.zip', index_col='ID')
sample_submission = pd.read_csv('sample_submission.csv.zip')
```

```python
print(f'train.shape = {train.shape}\n')
train.head(2)
```

```python
print(f'test.shape = {test.shape}\n')
test.head(2)
```

```python
print(f'sample_submission.shape = {sample_submission.shape}\n')
sample_submission.head(2)
```

## 2. Train valid split

```python
RANDOM_SEED = 42
```

```python
X_train, X_valid = train_test_split(
    train, test_size=0.2, random_state=RANDOM_SEED,
)
X_train.shape, X_valid.shape
```

## 3. Set up environment in PyCaret

```python
s = setup(
    data = X_train, target = 'y',  
    imputation_type = 'iterative',
    categorical_iterative_imputer = 'lightgbm',
    handle_unknown_categorical = True, 
    unknown_categorical_method = 'most_frequent',    
    remove_outliers = True,
    normalize = True,
    normalize_method = 'robust',    
    transformation = True,
    feature_selection = True,
    combine_rare_levels = True,    
    ignore_low_variance = True,
    use_gpu = True
)
```

## 4. Compare all models

```python
top3 = compare_models(n_select = 3, sort = 'r2')
```

```python
best = top3[0]
evaluate_model(best)
```

```python
predict_model(best)
```

## 5. Tune a Model

```python
tuned_best = tune_model(best, n_iter = 25)
```

```python
predict_model(tuned_best);
```

## 6. Ensemble a Model

```python
blender_top3 = blend_models([tuned_best, top3[1], top3[2]])
```

```python
stacker = stack_models([tuned_best, top3[1], top3[2]])
```

## 7. Finalize Model for Deployment

```python
final_best = finalize_model(blender_top3)
predict_model(final_best)
```

## 8. Predict on Unseen Data

```python
unseen_predictions = predict_model(final_best, data=X_valid.drop(['y'], axis=1))
unseen_predictions.head()
```

```python
print('Valid R2: ', round(r2_score(X_valid['y'], unseen_predictions['Label']), 4))
```

## 9. Save / load the Model

```python
save_model(final_best, 'my_final_best_model')
```

```python
loaded_model = load_model('my_final_best_model')
```

```python
new_prediction = predict_model(loaded_model, data=X_valid.drop(['y'], axis=1))
```

```python
new_prediction.head()
```

```python
print('Valid R2: ', round(r2_score(X_valid['y'], new_prediction['Label']), 4))
```

## 10. Final prediction

```python
predicts = predict_model(loaded_model, data=test)
```

```python
submission = predicts['Label'].reset_index()
submission.columns = ['ID', 'y']
submission.tail(2)
```
